# K Means Image Segmentation

### Segmentation With Lower K Value

```python
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image

def k_means(image, k=3, max_iterations=100):
    # Flatten the image to a 2D array of RGB values
    pixels = np.array(image).reshape((-1, 3))

    # Randomly initialize cluster centers
    centers = pixels[np.random.choice(pixels.shape[0], k, replace=False)]

    for iteration in range(max_iterations):
        print("Iteration: ", iteration)
        # Assign each pixel to the nearest cluster
        distances = np.linalg.norm(pixels - centers[:, np.newaxis], axis=2)
        labels = np.argmin(distances, axis=0)

        # Update cluster centers
        for i in range(k):
            centers[i] = np.mean(pixels[labels == i], axis=0)

    # Map each pixel to its cluster center
    segmented_image = centers[labels].reshape(np.array(image).shape)

    return segmented_image.astype(np.uint8)

# Load an example image
image_path = "./me.jpeg"
image = Image.open(image_path)

kValue = 6
segmented_image = k_means(image, k=kValue)

plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
plt.title("Original Image")
plt.imshow(image)
plt.axis("off")

plt.subplot(1, 2, 2)
plt.title("Segmented Image (K=" + str(kValue) + ")")
plt.imshow(segmented_image)
plt.axis("off")

plt.show()
```
